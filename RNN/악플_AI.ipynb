```python
import urllib.request
urllib.request.urlretrieve('https://raw.githubusercontent.com/bab2min/corpus/master/sentiment/naver_shopping.txt', 'shopping.txt')
```

```python
import pandas as pd
import numpy as np
raw = pd.read_table('shopping.txt',names=['rating','review'])
raw
```

```python
raw['label'] = np.where(raw['rating']>3,1,0)
raw
```

```python
from numpy._core.fromnumeric import sort
# 데이터 이쁘게 하기

raw['review'] = raw['review'].astype(str).str.replace(r'[^ㄱ-하-ㅣ가-힣0-9 ]','', regex=True)

# print(raw.isnull().sum())

raw.drop_duplicates( subset=['review'], inplace=True)
raw.head()

#bag of words
유니크문자 = raw['review'].tolist()
유니크문자 = ''.join(유니크문자)
유니크문자 = list(set(유니크문자))
유니크문자.sort()
유니크문자
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(char_level = True, oov_token='<OOV>')

문자리스트 = raw['review'].tolist()
tokenizer.fit_on_texts(문자리스트)

print(tokenizer.word_index)
print(문자리스트[0:10])

train_seq = tokenizer.texts_to_sequences(문자리스트)
print(train_seq[0:10])

Y = raw['label'].tolist()
print(Y[0:10])
```

```python
raw['length'] = raw['review'].str.len()

print(raw.describe())
print(raw.head())


raw['length'][raw['length']<100].count()
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

X = pad_sequences(train_seq, maxlen=100)

from  sklearn.model_selection import train_test_split
trainX, valX, trainY, valY = train_test_split(X, Y, test_size=0.2, random_state=42)

print(len(trainX))
print(len(valX))
```

```python
import tensorflow as tf
import numpy as np

max_len = trainX.shape[1]
vocab_size = len(tokenizer.word_index) + 1

EMB_DIM = 64

inputs = tf.keras.Input(shape=(max_len,), dtype="int32")
x = tf.keras.layers.Embedding(
    input_dim=vocab_size,
    output_dim=EMB_DIM,
    input_length=max_len
)(inputs)

# 문맥 학습 (순수 LSTM)
x = tf.keras.layers.Bidirectional(
    tf.keras.layers.LSTM(64, return_sequences=True)
)(x)

x = tf.keras.layers.Bidirectional(
    tf.keras.layers.LSTM(32)
)(x)

x = tf.keras.layers.Dropout(0.3)(x)
x = tf.keras.layers.Dense(64, activation="relu")(x)
x = tf.keras.layers.Dropout(0.2)(x)

# 정상=1, 악플=0 → sigmoid 출력은 "정상 확률"
outputs = tf.keras.layers.Dense(1, activation="sigmoid")(x)

model = tf.keras.Model(inputs, outputs)
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss="binary_crossentropy",
    metrics=[
        "accuracy",
        tf.keras.metrics.AUC(name="auc"),
    ],
)

model.summary()
```

```python
callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor="val_auc",
        patience=3,
        mode="max",
        restore_best_weights=True
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor="val_auc",
        factor=0.5,
        patience=2,
        mode="max",
        min_lr=1e-6
    ),
]
# numpy 타입으로 명확히
trainX_np = np.asarray(trainX, dtype=np.int32)
valX_np   = np.asarray(valX,   dtype=np.int32)

trainY_np = np.asarray(trainY, dtype=np.float32).reshape(-1, 1)
valY_np   = np.asarray(valY,   dtype=np.float32).reshape(-1, 1)

history = model.fit(
    trainX_np, trainY_np,
    validation_data=(valX_np, valY_np),
    epochs=15,
    batch_size=128,
    callbacks=callbacks,
    verbose=1
)
```

```python
import re
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences

def predict_only(text: str):
    # 전처리 (훈련 때랑 동일하게)
    text = re.sub(r'[^ㄱ-하-ㅣ가-힣0-9 ]', '', str(text))
    seq = tokenizer.texts_to_sequences([text])
    seq = pad_sequences(seq, maxlen=max_len, padding="post", truncating="post")

    p_normal = float(model.predict(seq, verbose=0)[0][0])  # 정상 확률(라벨이 정상=1 이니까)
    p_toxic  = 1.0 - p_normal                              # 악플 확률

    return {
        "text": text,
        "p_normal": p_normal,
        "p_toxic": p_toxic,
        "pred": "정상" if p_normal >= 0.5 else "악플"  # 필요 없으면 이 줄 지워
    }

print(predict_only("너 진짜 ㅈㄴ 못한다"))
print(predict_only("좋은 하루 보내세요"))
```
